```python
import pandas as pd
df = pd.read_json("dataset.json", orient="split")
#df=df.resample('3h').mean()
df_dir = df[['Direction']]
df_dir = df_dir[df_dir.Direction.notnull()]
print(df_dir)
```

```python
df=df.resample('3h').mean()
print(df)
```

```python
#df_full = pd.merge(df, df_dir, left_index=True, right_index=True)
df_full = df.join(df_dir)
#df_full = pd.concat([df, df_dir], axis=1)
df_full = df_full.dropna()
df_full = df_full[['Total','Speed','Direction']]
print(df_full.to_string())
```
